## Variáveis: 

- User_ID: Código do respondente
- Age: Idade do respondente
- Gender: Genêro (Female, Male, Other)
- Tecnology_Usage_Hours: Uso de tecnologia diária em horas por dia
- Gaming_Hours: Tempo em horas de "gameplay" por dia
- Screen_Time_Hours: Tempo em horas de uso de tela por dia
- Mental_Health_Status: Status da saúde mental (Excellent, Good, Fair, Poor)
- Stress_Level: Nível de estresse (Medium, Low, High)
- Sleep_Hours: Tempo de sono em horas por dia
- Physical_Activity_Hours: Tempo de atividade física em horas por dia
- Support_Systems_Access: Se há acesso à sistemas de suporte (Yes, No)
- Work_Environment_Impact: Impacto envolvido no ambiente de trabalho (Negative, Neutral, Positive)
- Online_Support_Usage: Uso de suporte online (Yes, No

#### Esse conjunto de dados oferece insights sobre como o uso diário da tecnologia, incluindo mídias sociais e tempo de tela, afeta a saúde mental. Ele captura vários padrões comportamentais e suas correlações com indicadores de saúde mental, como níveis de estresse, qualidade do sono e produtividade.

Este trabalho tem como objetivo realizar uma análise abrangente sobre os dados relacionados à saúde mental e ao uso de tecnologias, com base na base de dados “Mental Health and Technology Usage 2024”. Inicialmente, será feita uma análise descritiva, a fim de explorar e resumir as principais características das variáveis envolvidas. Em seguida, será conduzida uma análise fatorial, com o intuito de identificar possíveis dimensões subjacentes aos dados, reduzindo a complexidade do conjunto de variáveis. Posteriormente, será aplicada uma análise de conglomerados (cluster analysis) para agrupar os indivíduos com base em suas semelhanças, permitindo identificar perfis distintos entre os participantes. Por fim, será desenvolvida uma regressão logística, com o objetivo de modelar a probabilidade de ocorrência de desfechos relacionados à saúde mental com base em variáveis explicativas selecionadas. Essas etapas permitirão uma compreensão mais profunda dos padrões e relações existentes na amostra analisada.

### Relação entre "Screen_Time_Hours" e "Stress_Level"

```python
# Criar as categorias para Screen_Time_Hours
bins = [0, 3, 6, 9, float('inf')]  # Faixas de tempo de tela
labels = ["< 3h", "3-6h", "6-9h", "> 9h"]  # Nomes das faixas
df['Screen_Time_Category'] = pd.cut(df['Screen_Time_Hours'], bins=bins, labels=labels, right=False)

# Contar quantas pessoas em cada categoria de tela possuem cada nível de estresse
stress_counts = df.groupby(['Screen_Time_Category', 'Stress_Level'], observed = False).size().reset_index(name='Count')
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=stress_counts, x='Screen_Time_Category', y='Count', hue='Stress_Level', palette='coolwarm')

# Configurações do gráfico
plt.title("Distribuição do Nível de Estresse por Tempo de Tela")
plt.xlabel("Tempo de Tela (Horas)")
plt.ylabel("Quantidade de Pessoas")
plt.legend(title="Nível de Estresse")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```

```python
# Criar o gráfico de caixa
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Stress_Level', y='Screen_Time_Hours', hue='Stress_Level')

# Configurações do gráfico
plt.title("Relação entre Tempo de Tela e Nível de Estresse")
plt.xlabel("Nível de Estresse")
plt.ylabel("Horas de Tela")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Relação entre "Physical_Activity_Hours" e "Stress_Level"

```python
# Criar as categorias para Physical_Activity_Hours
bins = [0, 1, 3, 6, float('inf')]  # Faixas de atividade física
labels = ["0-1h (Sedentário)", "1-3h (Leve)", "3-6h (Moderado)", "> 6h (Intenso)"]  # Nomes das faixas
df['Physical_Activity_Category'] = pd.cut(df['Physical_Activity_Hours'], bins=bins, labels=labels, right=False)

# Contar quantas pessoas em cada categoria de atividade física possuem cada nível de estresse
stress_counts = df.groupby(['Physical_Activity_Category', 'Stress_Level'], observed = False).size().reset_index(name='Count')
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=stress_counts, x='Physical_Activity_Category', y='Count', hue='Stress_Level', palette='coolwarm')

# Configurações do gráfico
plt.title("Distribuição do Nível de Estresse por Tempo de Atividade Física")
plt.xlabel("Tempo de Atividade Física (Horas)")
plt.ylabel("Quantidade de Pessoas")
plt.legend(title="Nível de Estresse")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```
