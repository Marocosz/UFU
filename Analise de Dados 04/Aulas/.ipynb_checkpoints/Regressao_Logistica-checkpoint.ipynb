### A Regressão Logística:

> - Uma variável dependente binária e variáveis independentes métricas e não métricas: **Regressão Logística Binária**
> - Uma variável dependente com mais de duas categorias e variáveis independentes métricas e não métricas: **Regressão Logística Multinomial**
> - **Binária:** A variável dependente representa grupos binários (homem/mulher, sucesso/fracasso, bom/mau, compra/não compra, etc.). Um grupo é representado por 1 e o outro é representado por 0.
> - A regressão logística foi elaborada para prever a probabilidade de um evento ocorrer




### A Função Logística

> - Um modelo logístico segue a seguinte equação:

<FONT SIZE="4">
$\begin{aligned}
f(z) = \frac{1} {1 + e^{(-z)}}
 \end{aligned}$
</FONT>
<br>

#### Sendo Z

<FONT SIZE="4">
$\begin{aligned}
Z = \ln(\frac{p} {1 - p}) = \alpha + \beta_1X_1 + \beta_2X_2 + ... + \beta_kX_k
 \end{aligned}$
</FONT>
<br>

- 𝑝 indica a probabilidade de ocorrência de determinado evento de interesse, 𝑋 representa o vetor de variáveis explicativas (independentes) e 𝛼 e 𝛽 os parâmetros do modelo.

- O termo: <FONT SIZE="4">
$\begin{aligned}
\ln(\frac{p} {1 - p}) 
 \end{aligned}$
</FONT> é chamado de <b>logit</b>

- O termo: <FONT SIZE="4">
$\begin{aligned}
\frac{p} {1 - p}
 \end{aligned}$
</FONT> representa a <b>chance (odds)</b> de ocorrência do evento de interesse


### A Curva Logística

Os valores se situam entre 0 e 1, representando a probabilidade de ocorrência do evento de interesse.

### Premissas

> - As variáveis independentes podem ser intervalares, razão ou dicotômicas (binárias);
> - Relação linear entre o vetor das variáveis explicativas (independentes) 𝑋 e a variável dependente 𝑌;
> - Valor esperado (𝐸) dos resíduos é igual a zero
> - Ausência de autocorrelação
> - Ausência de correlação entre os resíduos e as variáveis explicativas
> - Ausência de multicolinearidade
> - A normalidade e igualdade da matriz de variância-covariância entre os grupos são dispensados.




### Seleção de Variáveis Dependente e Independentes:

- Variável Dependente:
> - O número de grupos (categorias) pode ser dois ou mais
>> - Duas categorias: Regressão Logística Binária
>> - Três categorias ou mais: Regressão Logística Multinomial
> - Devem ser mutuamente excludentes e cobrir todos os casos (exaustivo).
> - Caso a variável dependente seja métrica, pode-se transformá-la em uma variável categórica
- Variáveis Independentes
> - Podem ser métricas ou não métricas


### Divisão da Amostra:

- É muito comum um estudo de validação do modelo, usando recursos de divisão da amostra  - Partição da amostra ou validação cruzada (crossvalidation); 
- Este procedimento consiste em dividir a amostra em duas partes: 
> - Modelagem (50 a 75%) e 
> - Teste (50 a 25%).
- Se a amostra for muito pequena, validar a função no mesmo grupo que foi utilizado para desenvolver a função.


### Medidas de Ajuste do Modelo

- Valor de verossimilhança – 2LL ou -2log verossimilhança –Quanto menor o valor de -2LL, melhor é a adequação do modelo. Quando a verossimilhança for 1, indicando ajuste perfeito, o valor de -2LL é zero.

- <FONT SIZE="4">
$\begin{aligned}
\text{Pseudo } R^2: R^2logit = \frac{-2LL_0 -(-2LL_\beta)} {-2LL_0}
 \end{aligned}$
</FONT>

- <FONT SIZE="4">
$\begin{aligned}
\text{Cox & Snell } R^2 \text{(semelhante ao } R^2 \text{(da regressão linear múltipla): } R^2_{CS} = 1 - (\frac{L_0} {L_\beta})^\frac{2}{N} \therefore R^2_{CS MAX} = 1 - (L_0)^\frac{2}{N}
 \end{aligned}$
</FONT>

- <FONT SIZE="4">
$\begin{aligned}
\text{Nagelkerke  } R^2 \text{: }  R^2_N = \frac{R^2_{CS}} {R^2_{CS MAX}}
 \end{aligned}$
</FONT>

#### Criação de Variáveis Dummy

No SPSS as variáveis categóricas podem ser utilizadas como variáveis independentes da Rergressão Logística, sem a preocupação de transformá-las. No próprio algorítmo, o SPSS faz o tratamento. Contudo, aqui, é necessário trasformar as variáveis categóricas em **Variáveis Dummy** previamente, para que possam ser utilizadas na regressão logística<br>

<p> Função Utilizada: <br>
<a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html">get_dummies()</a> <br>
</p>


```python
cat_vars=['x1','x2','x3','x4','x5','x23']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df[var], prefix=var)
    df1=df.join(cat_list)
    df=df1
cat_vars=['x1','x2','x3','x4','x5','x23']
data_vars=df.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]
```

#### Over-Sampling - Balanceamento da amostra

- Utilização do Algoritmo SMOTE (Synthetic Minority Oversampling Technique). 
> - Cria amostras sintéticas a partir da categoria de menor tamanho. Não é copia.


<p> Função Utilizada: <br>
<a href="http://imbalanced-learn.org/en/stable/over_sampling.html#smote-variants">over_sampling()</a> <br>
</p>


```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import pandas as pd

X = df2.loc[:, df2.columns != 'x4_num']
y = df2.loc[:, df2.columns == 'x4_num']

os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
columns = X_train.columns

# Utilizando o método correto 'fit_resample'
os_data_X, os_data_y = os.fit_resample(X_train, y_train)

os_data_X = pd.DataFrame(data=os_data_X, columns=columns)
os_data_y = pd.DataFrame(data=os_data_y, columns=['x4_num'])

# Verificando os resultados
print("A Nova Amostra Tem Tamanho Igual a: ", len(os_data_X))
print("Número de Clientes dos EUA/América do Norte: ", len(os_data_y[os_data_y['x4_num'] == 0]))
print("Número de Clientes de Fora da América do Norte", len(os_data_y[os_data_y['x4_num'] == 1]))
print("Proporção de Clientes dos EUA/América do Norte na Nova Amostra: ", len(os_data_y[os_data_y['x4_num'] == 0]) / len(os_data_X))
print("Proporção de Clientes de Fora da América do Norte na Nova Amostra: ", len(os_data_y[os_data_y['x4_num'] == 1]) / len(os_data_X))
```

### Recursive Feature Elimination - Seleção de Variáveis

A eliminação recursiva de variáveis (RFE) é baseda na ideia da construção de modelos repetidamente e escolher as variáveis como melhor performance. O processo é aplicado a todas as variáveis exaustivamente.

```python
os_data_X.head(3)
```

<p> Função Utilizada: <br>
<a href="https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html">RFE()</a> <br>
</p>


```python
df2_vars=df2.columns.values.tolist()
y=['x4_num']
X=[i for i in df2_vars if i not in y]
X
```

```python
df2_vars=df2.columns.values.tolist()
y=['x4_num']
X=[i for i in df2_vars if i not in y]


logreg = LogisticRegression()
rfe = RFE(logreg)
rfe = rfe.fit(os_data_X, os_data_y.values.ravel())
print(rfe.support_)
print(rfe.ranking_)
```

##### Análise RFE

Pela RFE, estas seriam as variáveis a serem incluídas no modelo. Contudo, ao incluir todas as variáveis, o modelo ficou super ajsutado, impossibilitando os cálculos matriciais, além de apresentar variáveis não significativas esatíticamente. <br>

['x6', 'x9', 'x10', 'x11', 'x12', 'x15', 'x17', 'x20', 'x21', 'x1_Entre 1 e 5 Anos', 'x1_Menos de 1 Ano','x3_Grande (500+)', 'x3_Pequena (0 to 499)'] <br>

Assim, após testes com as variáveis, a solução final ficou com variáveis reduzidas, conforme passo a seguir.

```python
variaveis=['x6', 'x10', 'x11', 'x12', 'x15', 'x17'] 

X=os_data_X[variaveis]
y=os_data_y['x4_num']
```

### Implementando o Modelo
