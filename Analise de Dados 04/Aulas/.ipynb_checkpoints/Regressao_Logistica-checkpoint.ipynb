### A Regress√£o Log√≠stica:

> - Uma vari√°vel dependente bin√°ria e vari√°veis independentes m√©tricas e n√£o m√©tricas: **Regress√£o Log√≠stica Bin√°ria**
> - Uma vari√°vel dependente com mais de duas categorias e vari√°veis independentes m√©tricas e n√£o m√©tricas: **Regress√£o Log√≠stica Multinomial**
> - **Bin√°ria:** A vari√°vel dependente representa grupos bin√°rios (homem/mulher, sucesso/fracasso, bom/mau, compra/n√£o compra, etc.). Um grupo √© representado por 1 e o outro √© representado por 0.
> - A regress√£o log√≠stica foi elaborada para prever a probabilidade de um evento ocorrer




### A Fun√ß√£o Log√≠stica

> - Um modelo log√≠stico segue a seguinte equa√ß√£o:

<FONT SIZE="4">
$\begin{aligned}
f(z) = \frac{1} {1 + e^{(-z)}}
 \end{aligned}$
</FONT>
<br>

#### Sendo Z

<FONT SIZE="4">
$\begin{aligned}
Z = \ln(\frac{p} {1 - p}) = \alpha + \beta_1X_1 + \beta_2X_2 + ... + \beta_kX_k
 \end{aligned}$
</FONT>
<br>

- ùëù indica a probabilidade de ocorr√™ncia de determinado evento de interesse, ùëã representa o vetor de vari√°veis explicativas (independentes) e ùõº e ùõΩ os par√¢metros do modelo.

- O termo: <FONT SIZE="4">
$\begin{aligned}
\ln(\frac{p} {1 - p}) 
 \end{aligned}$
</FONT> √© chamado de <b>logit</b>

- O termo: <FONT SIZE="4">
$\begin{aligned}
\frac{p} {1 - p}
 \end{aligned}$
</FONT> representa a <b>chance (odds)</b> de ocorr√™ncia do evento de interesse


### A Curva Log√≠stica

Os valores se situam entre 0 e 1, representando a probabilidade de ocorr√™ncia do evento de interesse.

### Premissas

> - As vari√°veis independentes podem ser intervalares, raz√£o ou dicot√¥micas (bin√°rias);
> - Rela√ß√£o linear entre o vetor das vari√°veis explicativas (independentes) ùëã e a vari√°vel dependente ùëå;
> - Valor esperado (ùê∏) dos res√≠duos √© igual a zero
> - Aus√™ncia de autocorrela√ß√£o
> - Aus√™ncia de correla√ß√£o entre os res√≠duos e as vari√°veis explicativas
> - Aus√™ncia de multicolinearidade
> - A normalidade e igualdade da matriz de vari√¢ncia-covari√¢ncia entre os grupos s√£o dispensados.




### Sele√ß√£o de Vari√°veis Dependente e Independentes:

- Vari√°vel Dependente:
> - O n√∫mero de grupos (categorias) pode ser dois ou mais
>> - Duas categorias: Regress√£o Log√≠stica Bin√°ria
>> - Tr√™s categorias ou mais: Regress√£o Log√≠stica Multinomial
> - Devem ser mutuamente excludentes e cobrir todos os casos (exaustivo).
> - Caso a vari√°vel dependente seja m√©trica, pode-se transform√°-la em uma vari√°vel categ√≥rica
- Vari√°veis Independentes
> - Podem ser m√©tricas ou n√£o m√©tricas


### Divis√£o da Amostra:

- √â muito comum um estudo de valida√ß√£o do modelo, usando recursos de divis√£o da amostra  - Parti√ß√£o da amostra ou valida√ß√£o cruzada (crossvalidation); 
- Este procedimento consiste em dividir a amostra em duas partes: 
> - Modelagem (50 a 75%) e 
> - Teste (50 a 25%).
- Se a amostra for muito pequena, validar a fun√ß√£o no mesmo grupo que foi utilizado para desenvolver a fun√ß√£o.


### Medidas de Ajuste do Modelo

- Valor de verossimilhan√ßa ‚Äì 2LL ou -2log verossimilhan√ßa ‚ÄìQuanto menor o valor de -2LL, melhor √© a adequa√ß√£o do modelo. Quando a verossimilhan√ßa for 1, indicando ajuste perfeito, o valor de -2LL √© zero.

- <FONT SIZE="4">
$\begin{aligned}
\text{Pseudo } R^2: R^2logit = \frac{-2LL_0 -(-2LL_\beta)} {-2LL_0}
 \end{aligned}$
</FONT>

- <FONT SIZE="4">
$\begin{aligned}
\text{Cox & Snell } R^2 \text{(semelhante ao } R^2 \text{(da regress√£o linear m√∫ltipla): } R^2_{CS} = 1 - (\frac{L_0} {L_\beta})^\frac{2}{N} \therefore R^2_{CS MAX} = 1 - (L_0)^\frac{2}{N}
 \end{aligned}$
</FONT>

- <FONT SIZE="4">
$\begin{aligned}
\text{Nagelkerke  } R^2 \text{: }  R^2_N = \frac{R^2_{CS}} {R^2_{CS MAX}}
 \end{aligned}$
</FONT>

#### Cria√ß√£o de Vari√°veis Dummy

No SPSS as vari√°veis categ√≥ricas podem ser utilizadas como vari√°veis independentes da Rergress√£o Log√≠stica, sem a preocupa√ß√£o de transform√°-las. #### Over-Sampling - Balanceamento da amostra

- Utiliza√ß√£o do Algoritmo SMOTE (Synthetic Minority Oversampling Technique). 
> - Cria amostras sint√©ticas a partir da categoria de menor tamanho. N√£o √© copia.


<p> Fun√ß√£o Utilizada: <br>
<a href="http://imbalanced-learn.org/en/stable/over_sampling.html#smote-variants">over_sampling()</a> <br>
</p>


```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import pandas as pd

X = df2.loc[:, df2.columns != 'x4_num']
y = df2.loc[:, df2.columns == 'x4_num']

os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
columns = X_train.columns

# Utilizando o m√©todo correto 'fit_resample'
os_data_X, os_data_y = os.fit_resample(X_train, y_train)

os_data_X = pd.DataFrame(data=os_data_X, columns=columns)
os_data_y = pd.DataFrame(data=os_data_y, columns=['x4_num'])

# Verificando os resultados
print("A Nova Amostra Tem Tamanho Igual a: ", len(os_data_X))
print("N√∫mero de Clientes dos EUA/Am√©rica do Norte: ", len(os_data_y[os_data_y['x4_num'] == 0]))
print("N√∫mero de Clientes de Fora da Am√©rica do Norte", len(os_data_y[os_data_y['x4_num'] == 1]))
print("Propor√ß√£o de Clientes dos EUA/Am√©rica do Norte na Nova Amostra: ", len(os_data_y[os_data_y['x4_num'] == 0]) / len(os_data_X))
print("Propor√ß√£o de Clientes de Fora da Am√©rica do Norte na Nova Amostra: ", len(os_data_y[os_data_y['x4_num'] == 1]) / len(os_data_X))
```
