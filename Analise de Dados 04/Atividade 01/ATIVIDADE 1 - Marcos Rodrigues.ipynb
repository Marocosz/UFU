```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Carregar os dados
df = pd.read_excel("PEDIDOS_INDUSTRIA_TEXTIL_V1.xlsx", sheet_name="BASE")
```

```python
df[['QTDVENDIDA','VLRVENDIDO', 'QTDFATURADA', 'VLRFATURADO']].describe()
```

```python
# Converter datas
df["MESANOVENDA"] = pd.to_datetime(df["MESANOVENDA"])
```

```python
agrupado_por_mes = df.groupby(df["MESANOVENDA"].dt.to_period("M")).sum(numeric_only=True)[["QTDVENDIDA", "VLRVENDIDO", "QTDFATURADA", "VLRFATURADO"]]
```

```python
# Criar gráfico de barras
plt.figure(figsize=(12, 6))
agrupado_por_mes.plot(kind="bar", figsize=(12, 6), colormap="viridis")
plt.title("Soma das Vendas e Faturamento por Mês")
plt.xlabel("Mês")
plt.ylabel("Valores")
plt.xticks(rotation=45)
plt.legend(title="Indicadores")
plt.grid(axis="y")
plt.show()
```
